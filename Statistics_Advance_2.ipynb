 1. What is hypothesis testing in statistics?


   - Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves formulating a null hypothesis (H_0) and an alternative hypothesis (H_1), collecting data, and using statistical tests to determine whether there is enough evidence to reject the null hypothesis.

 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?


   - The null hypothesis (H_0) is a statement of no effect, no difference, or no relationship. It represents the status quo or a commonly held belief. The alternative hypothesis (H_1 or H_a) is a statement that contradicts the null hypothesis. It represents what the researcher is trying to prove or what they believe to be true. The goal of hypothesis testing is to see if there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

 3.  What is the significance level in hypothesis testing, and why is it important?


   - The significance level (\alpha) is the probability of rejecting the null hypothesis when it is actually true (i.e., committing a Type I error). Common significance levels are 0.05 (5%), 0.01 (1%), or 0.10 (10%). It is important because it sets the threshold for how much evidence is needed to reject the null hypothesis. A smaller \alpha means you require stronger evidence to reject H_0.

 4. What does a P-value represent in hypothesis testing?


   - The P-value (probability value) is the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming that the null hypothesis is true. It quantifies the strength of evidence against the null hypothesis.


 5. How do you interpret the P-value in hypothesis testing?


   * If P-value \le \alpha: You reject the null hypothesis. This suggests that the observed data is unlikely if the null hypothesis were true, providing evidence in favor of the alternative hypothesis.
   * If P-value > \alpha: You fail to reject the null hypothesis. This suggests that the observed data is not unusual if the null hypothesis were true, meaning there isn't enough evidence to support the alternative hypothesis.


 6. What are Type I and Type II errors in hypothesis testing?


   * Type I error (\alpha): Rejecting the null hypothesis when it is actually true (a "false positive").
   * Type II error (\beta): Failing to reject the null hypothesis when it is actually false (a "false negative").


 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?


   * One-tailed test (directional): The alternative hypothesis specifies a direction (e.g., the mean is greater than, or less than, a certain value). The critical region for rejecting H_0 is entirely in one tail of the distribution.
   * Two-tailed test (non-directional): The alternative hypothesis does not specify a direction (e.g., the mean is simply different from a certain value). The critical region is split between both tails of the distribution.


 8. What is the Z-test, and when is it used in hypothesis testing?


   - The Z-test is a statistical hypothesis test used to determine if there is a significant difference between a sample mean and a population mean, or between two sample means, when the population standard deviation is known and the sample size is large (typically n \ge 30). It assumes that the data are normally distributed.


 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?


   - The Z-score (or standard score) measures how many standard deviations an element is from the mean.
   For a single sample mean: Z = (\bar{x} - \mu) / (\sigma / \sqrt{n})
   Where:
   * \bar{x} = sample mean
   * \mu = population mean
   * \sigma = population standard deviation
   * n = sample size
     It represents the position of a data point relative to the mean of a distribution, standardized by the standard deviation. In hypothesis testing, it indicates how far the sample mean is from the hypothesized population mean in terms of standard errors.


 10. What is the T-distribution, and when should it be used instead of the normal distribution?


   The T-distribution (Student's t-distribution) is a probability distribution that arises when estimating the mean of a normally distributed population when the sample size is small and the population standard deviation is unknown.
   It should be used instead of the normal distribution when:
   * The sample size (n) is small (typically n < 30).
   * The population standard deviation (\sigma) is unknown and is estimated from the sample standard deviation (s).


 11. What is the difference between a Z-test and a T-test?


   The main difference lies in when they are used:
   * Z-test: Used when the population standard deviation (\sigma) is known, or when the sample size (n) is large (typically n \ge 30), allowing the sample standard deviation to be a good estimate of the population standard deviation.
   * T-test: Used when the population standard deviation (\sigma) is unknown and the sample size (n) is small (typically n < 30). The T-distribution accounts for the increased uncertainty due to estimating \sigma from a small sample.


 12. What is the T-test, and how is it used in hypothesis testing?


  -  The T-test is a statistical hypothesis test used to compare means. It is used when the population standard deviation is unknown and/or the sample size is small.
   Common uses in hypothesis testing include:
   * One-sample T-test: Comparing a sample mean to a known population mean.
   * Independent samples T-test: Comparing the means of two independent groups.
   * Paired samples T-test: Comparing the means of two related (dependent) groups.


 13. What is the relationship between Z-test and T-test in hypothesis testing?


  -  The T-test is a generalization of the Z-test. As the sample size (n) increases, the T-distribution approaches the standard normal (Z) distribution. In essence, for large sample sizes, the T-test and Z-test yield very similar results because the sample standard deviation becomes a very good estimate of the population standard deviation.


 14. What is a confidence interval, and how is it used to interpret statistical results?


  -  A confidence interval (CI) is a range of values within which the true population parameter (e.g., mean, proportion) is estimated to lie with a certain level of confidence. For example, a 95% CI means that if you were to repeat the sampling process many times, 95% of the constructed intervals would contain the true population parameter.
   It is used to interpret statistical results by providing a plausible range for the true parameter, complementing the point estimate. If a hypothesized value (e.g., the null hypothesis mean) falls outside the confidence interval, it provides evidence to reject the null hypothesis.


 15.  What is the margin of error, and how does it affect the confidence interval?


  -  The margin of error (MOE) is the "half-width" of the confidence interval. It quantifies the maximum expected difference between the sample estimate and the true population parameter due to random sampling variability.
   The confidence interval is calculated as: Point Estimate \pm Margin of Error.
   A larger margin of error results in a wider confidence interval, indicating less precision in the estimate. A smaller margin of error results in a narrower confidence interval, indicating greater precision.


 16.  How is Bayes' Theorem used in statistics, and what is its significance?


  -  Bayes' Theorem is a fundamental theorem in probability that describes how to update the probability of a hypothesis based on new evidence. It relates the conditional probability of an event to its inverse.
   P(A|B) = [P(B|A) * P(A)] / P(B)
   Where:
   * P(A|B) = Posterior probability of hypothesis A given evidence B
   * P(B|A) = Likelihood of evidence B given hypothesis A
   * P(A) = Prior probability of hypothesis A
   * P(B) = Marginal probability of evidence B
     Its significance lies in its ability to integrate prior knowledge or beliefs with observed data to arrive at updated (posterior) probabilities. This is central to Bayesian statistics, which provides an alternative framework to frequentist statistics for inference and decision-making, particularly useful in situations with limited data or when incorporating expert opinion.


 17.  What is the Chi-square distribution, and when is it used?


 -   The Chi-square (\chi^2) distribution is a continuous probability distribution that arises in the context of sums of squared standard normal variates. It is primarily used in hypothesis testing for:
   * Goodness-of-fit tests: To determine if observed frequencies match expected frequencies in a categorical variable.
   * Tests of independence: To determine if there is a statistically significant association between two categorical variables in a contingency table.
   * Tests of homogeneity: To determine if the distribution of a categorical variable is the same across different populations.
   * Estimating population variance or standard deviation.


 18. What is the Chi-square goodness of fit test, and how is it applied?


  -  The Chi-square goodness of fit test is used to assess whether an observed frequency distribution of a categorical variable differs significantly from an expected frequency distribution.
   Application:
   * State hypotheses: H_0: The observed distribution fits the expected distribution. H_1: The observed distribution does not fit the expected distribution.
   * Calculate expected frequencies: Based on the null hypothesis.
   * Calculate the Chi-square test statistic: \chi^2 = \sum [(O_i - E_i)^2 / E_i], where O_i is the observed frequency and E_i is the expected frequency for each category.
   * Determine degrees of freedom: df = k - 1 - p, where k is the number of categories and p is the number of parameters estimated from the data (usually 0).
   * Compare \chi^2 statistic to critical value or P-value: If the calculated \chi^2 is greater than the critical value (or P-value \le \alpha), reject H_0.


 19. What is the F-distribution, and when is it used in hypothesis testing?


   - The F-distribution (Snedecor's F-distribution) is a continuous probability distribution that arises in the context of ratios of variances. It is primarily used in hypothesis testing for:
   * Analysis of Variance (ANOVA): To compare the means of three or more groups.
   * Testing the equality of variances between two or more populations.
   * Regression analysis: To test the overall significance of a regression model (e.g., F-test for overall model fit).


 20.  What is an ANOVA test, and what are its assumptions?


  -  ANOVA (Analysis of Variance) is a statistical hypothesis test used to determine if there are any statistically significant differences between the means of three or more independent groups. It works by partitioning the total variance in a dataset into different sources of variation.
   Assumptions of ANOVA:
   * Independence of observations: The observations within and between groups must be independent.
   * Normality: The data within each group should be approximately normally distributed. (ANOVA is relatively robust to minor deviations from normality, especially with larger sample sizes due to the Central Limit Theorem).
   * Homoscedasticity (Homogeneity of variances): The variances of the populations from which the samples are drawn should be equal. (Levene's test or Bartlett's test can be used to check this assumption).


 21.  What are the different types of ANOVA tests?The main types of ANOVA tests include:


   * One-Way ANOVA: Used when there is one categorical independent variable (factor) with three or more levels, and one continuous dependent variable.
   * Two-Way ANOVA: Used when there are two categorical independent variables (factors) and one continuous dependent variable. It can also assess interaction effects between the factors.
   * N-Way ANOVA (Multifactor ANOVA): An extension of two-way ANOVA to more than two independent categorical variables.
   * Repeated Measures ANOVA: Used when the same subjects are measured multiple times under different conditions or at different time points (dependent groups).
   * MANOVA (Multivariate Analysis of Variance): Used when there are one or more categorical independent variables and two or more continuous dependent variables.


 22. What is the F-test, and how do it relate to hypothesis testing?


  -  The F-test is a statistical test that uses the F-distribution to compare variances or to test the overall significance of models (like in ANOVA or regression).
   Relation to Hypothesis Testing:
   * In ANOVA, the F-test is used to test the null hypothesis that the means of all groups are equal. The F-statistic is the ratio of the variance between groups (mean square between) to the variance within groups (mean square within). A large F-value suggests that the variation between group means is greater than the variation within groups, leading to the rejection of the null hypothesis.
   * In regression analysis, the F-test is used to test the overall significance of the regression model, i.e., whether at least one of the independent variables significantly predicts the dependent variable.
   * In comparing variances, the F-test can be used to test the null hypothesis that two population variances are equal.

```python
import numpy as np
from scipy.stats import norm

# --- Input values ---
# Population parameters
population_mean = 100       # μ
population_std = 15         # σ (known population standard deviation)

# Sample data
sample = [102, 98, 101, 97, 100, 105, 99, 103, 104, 96]
sample_size = len(sample)
sample_mean = np.mean(sample)

# --- Z-test statistic formula ---
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# --- Calculate p-value (two-tailed) ---
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# --- Output ---
print(f"Sample Mean = {sample_mean}")
print(f"Z-Score = {z_score:.4f}")
print(f"P-Value = {p_value:.4f}")

# --- Interpretation ---
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")
```

```python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Output
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Hypothesis decision
    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference from the population mean.")

# Example Usage:
np.random.seed(42)
sample_data = np.random.normal(loc=105, scale=10, size=40)  # Simulated sample
population_mean = 100
population_std = 15

one_sample_z_test(sample_data, population_mean, population_std)
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Parameters ---
population_mean = 100       # μ
population_std = 15         # σ (known)
alpha = 0.05                # Significance level
sample_size = 40

# --- Simulate sample data ---
np.random.seed(42)
sample_data = np.random.normal(loc=105, scale=10, size=sample_size)
sample_mean = np.mean(sample_data)

# --- Z-test Calculation ---
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# --- Z critical value for two-tailed test ---
z_critical = norm.ppf(1 - alpha/2)

# --- Print results ---
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")
if abs(z_score) > z_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# --- Plotting the decision region ---
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution')

# Critical regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5)

# Sample Z-score line
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')

# Add critical value lines
plt.axvline(-z_critical, color='black', linestyle=':', label=f'Z-critical = ±{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle=':')

# Labels and legend
plt.title('Two-Tailed Z-Test Visualization')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(pop_mean, pop_std, sample_size, alt_mean, alpha=0.05):
    # Standard error
    se = pop_std / np.sqrt(sample_size)

    # Z-critical value for one-tailed test (right-tail)
    z_crit = norm.ppf(1 - alpha)

    # Critical value in original scale
    critical_value = pop_mean + z_crit * se

    # β: Probability of Type II error (fail to reject H0 when H1 is true)
    beta = norm.cdf((critical_value - alt_mean) / se)

    # Power = 1 - β
    power = 1 - beta

    # X range for plotting
    x = np.linspace(pop_mean - 4*se, alt_mean + 4*se, 1000)

    # Distributions
    h0_dist = norm.pdf(x, loc=pop_mean, scale=se)
    h1_dist = norm.pdf(x, loc=alt_mean, scale=se)

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(x, h0_dist, label='H₀ Distribution (Null)', color='blue')
    plt.plot(x, h1_dist, label='H₁ Distribution (Alternative)', color='green')

    # Type I error region
    plt.fill_between(x, h0_dist, where=(x > critical_value), color='red', alpha=0.4, label='Type I Error (α)')

    # Type II error region
    plt.fill_between(x, h1_dist, where=(x <= critical_value), color='orange', alpha=0.4, label='Type II Error (β)')

    # Critical value line
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print key values
    print(f"Critical Value (cutoff for rejecting H₀): {critical_value:.2f}")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the Test (1 - β): {power:.4f}")


visualize_type1_type2(
    pop_mean=100,       # H₀: μ = 100
    pop_std=15,         # σ
    sample_size=50,     # n
    alt_mean=105,       # H₁: μ = 105
    alpha=0.05          # significance level
)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

# --- Step 1: Simulate data for two independent samples ---
np.random.seed(0)
group1 = np.random.normal(loc=100, scale=10, size=30)  # Group 1: mean=100
group2 = np.random.normal(loc=105, scale=10, size=30)  # Group 2: mean=105

# --- Step 2: Perform independent two-sample T-test ---
t_stat, p_value = ttest_ind(group1, group2)

# --- Step 3: Display results ---
print("Group 1 Mean:", np.mean(group1))
print("Group 2 Mean:", np.mean(group2))
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# --- Step 4: Interpretation ---
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two groups.")
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# --- Step 1: Simulate paired data (e.g., before and after treatment) ---
np.random.seed(42)
before = np.random.normal(loc=70, scale=8, size=30)   # Before treatment
after = before + np.random.normal(loc=5, scale=5, size=30)  # After treatment (slightly improved)

# --- Step 2: Perform Paired T-Test ---
t_stat, p_value = ttest_rel(before, after)

# --- Step 3: Print results ---
print("Mean Before:", np.mean(before))
print("Mean After :", np.mean(after))
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference (paired).")
else:
    print("Fail to reject the null hypothesis: No significant difference (paired).")

# --- Step 4: Visualize comparison ---
plt.figure(figsize=(10, 5))

# Line plot for paired observations
for i in range(len(before)):
    plt.plot([0, 1], [before[i], after[i]], color='gray', linestyle='--', alpha=0.6)

# Mean points
plt.scatter(np.zeros(len(before)), before, color='blue', label='Before', alpha=0.7)
plt.scatter(np.ones(len(after)), after, color='green', label='After', alpha=0.7)

plt.xticks([0, 1], ['Before', 'After'])
plt.title('Paired Sample Comparison (Before vs After)')
plt.ylabel('Scores')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t, norm

def confidence_interval(data, confidence=0.95, population_std=None):
    """
    Calculate the confidence interval for a sample mean.

    Parameters
    ----------
    data : array-like
        Sample observations.
    confidence : float, optional
        Desired confidence level (default 0.95 for 95% CI).
    population_std : float or None, optional
        Known population standard deviation. If provided, uses Z-distribution;
        otherwise uses the t-distribution with df = n-1.

    Returns
    -------
    tuple (lower_bound, upper_bound)
    """
    data = np.asarray(data)
    n = len(data)
    mean = np.mean(data)

    if population_std is not None:
        # Z-based interval
        se = population_std / np.sqrt(n)
        z_crit = norm.ppf((1 + confidence) / 2)
        margin = z_crit * se
    else:
        # t-based interval
        sample_std = np.std(data, ddof=1)
        se = sample_std / np.sqrt(n)
        df = n - 1
        t_crit = t.ppf((1 + confidence) / 2, df)
        margin = t_crit * se

    return mean - margin, mean + margin

# --- Example usage ---
if __name__ == "__main__":
    np.random.seed(0)
    # simulate a small sample
    sample = np.random.normal(loc=50, scale=5, size=20)

    ci_lower, ci_upper = confidence_interval(sample, confidence=0.95)
    print(f"Sample mean: {np.mean(sample):.2f}")
    print(f"95% CI (t-based): [{ci_lower:.2f}, {ci_upper:.2f}]")

    # if you did know σ=5:
    ci_z = confidence_interval(sample, confidence=0.95, population_std=5)
    print(f"95% CI (Z-based): [{ci_z[0]:.2f}, {ci_z[1]:.2f}]")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t, norm

def margin_of_error(data, confidence=0.95, population_std=None):
    """
    Calculate the margin of error for a sample mean.

    Parameters
    ----------
    data : array-like
        Sample observations.
    confidence : float
        Desired confidence level (default is 0.95 for 95%).
    population_std : float or None
        If provided, uses Z-distribution; otherwise, uses t-distribution.

    Returns
    -------
    float
        Margin of error.
    """
    data = np.array(data)
    n = len(data)

    if population_std is not None:
        # Z-distribution
        se = population_std / np.sqrt(n)
        z_crit = norm.ppf((1 + confidence) / 2)
        moe = z_crit * se
    else:
        # T-distribution
        sample_std = np.std(data, ddof=1)
        se = sample_std / np.sqrt(n)
        t_crit = t.ppf((1 + confidence) / 2, df=n-1)
        moe = t_crit * se

    return moe

# --- Example Usage ---
if __name__ == "__main__":
    # Simulate a sample
    np.random.seed(1)
    sample_data = np.random.normal(loc=75, scale=12, size=25)

    moe = margin_of_error(sample_data, confidence=0.95)
    print(f"Sample Mean: {np.mean(sample_data):.2f}")
    print(f"Margin of Error (95% CI, t-based): ±{moe:.2f}")

    # With known population standard deviation (e.g., σ = 12)
    moe_z = margin_of_error(sample_data, confidence=0.95, population_std=12)
    print(f"Margin of Error (95% CI, Z-based): ±{moe_z:.2f}")
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior, likelihood, evidence):
    """
    Calculate the posterior probability using Bayes' theorem.

    Parameters:
    - prior: P(H), prior probability of hypothesis
    - likelihood: P(E|H), probability of evidence given hypothesis
    - evidence: P(E), total probability of evidence

    Returns:
    - posterior probability P(H|E)
    """
    return (likelihood * prior) / evidence


# Given data:
P_disease = 0.01          # Prior probability of disease
P_no_disease = 1 - P_disease

P_pos_given_disease = 0.99    # True positive rate
P_pos_given_no_disease = 0.05 # False positive rate

# Calculate total probability of testing positive (evidence)
P_pos = (P_pos_given_disease * P_disease) + (P_pos_given_no_disease * P_no_disease)

# Calculate posterior probability: P(disease | positive test)
posterior = bayes_theorem(P_disease, P_pos_given_disease, P_pos)

print(f"Probability of having disease given a positive test: {posterior:.4f} ({posterior*100:.2f}%)")
```

```python
#13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test of independence.

    Parameters:
    observed (2D array-like): Contingency table of observed frequencies

    Returns:
    numpy.ndarray: Expected frequencies
    """
    observed = np.array(observed)

    # Row totals and column totals
    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)
    total = observed.sum()

    # Expected frequencies = (row total * column total) / grand total
    expected = (row_totals @ col_totals) / total
    return expected

# Example observed data
observed_data = np.array([[30, 10],
                          [20, 20]])

expected_freq = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:\n", observed_data)
print("Expected Frequencies:\n", expected_freq)
```

```python
#16.Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare variances of two independent samples.

    Parameters:
    - sample1, sample2: arrays or lists of sample data
    - alpha: significance level (default 0.05)

    Returns:
    - F statistic
    - p-value for two-tailed test
    - interpretation string
    """
    n1 = len(sample1)
    n2 = len(sample2)

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Ensure var1 is the larger variance
    if var1 >= var2:
        F = var1 / var2
        dfn, dfd = n1 - 1, n2 - 1
    else:
        F = var2 / var1
        dfn, dfd = n2 - 1, n1 - 1

    # Two-tailed p-value calculation
    p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

    result = "Fail to reject null hypothesis: variances are equal." if p_value > alpha else \
             "Reject null hypothesis: variances are significantly different."

    return F, p_value, result

# --- Example Usage ---
if __name__ == "__main__":
    np.random.seed(0)
    # Generate two random samples with different variances
    sample_a = np.random.normal(loc=10, scale=5, size=30)
    sample_b = np.random.normal(loc=10, scale=8, size=35)

    F_stat, p_val, interpretation = f_test(sample_a, sample_b)

    print(f"F-statistic: {F_stat:.4f}")
    print(f"P-value: {p_val:.4f}")
    print(interpretation)
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups, alpha=0.05):
    """
    Check ANOVA assumptions: normality and equal variance for multiple groups.

    Parameters:
    - groups: variable number of array-like sample groups
    - alpha: significance level (default 0.05)

    Returns:
    - dict with results for normality and equal variance tests
    """
    results = {}

    # 1. Normality test per group using Shapiro-Wilk
    normality_results = {}
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        normality_results[f'Group {i}'] = {
            'W-statistic': stat,
            'p-value': p,
            'Normal': p > alpha
        }
    results['Normality'] = normality_results

    # 2. Equal variances test using Levene’s test (more robust than Bartlett)
    stat, p = levene(*groups)
    results['Equal Variance'] = {
        'Levene_statistic': stat,
        'p-value': p,
        'Equal Variance': p > alpha
    }

    # 3. Independence assumption:
    # Usually ensured by proper experimental design (random sampling, no repeated measures).
    results['Independence'] = 'Assumed based on study design (cannot be tested statistically)'

    return results


# --- Example usage ---
if __name__ == "__main__":
    import numpy as np

    np.random.seed(42)
    group1 = np.random.normal(20, 5, 30)
    group2 = np.random.normal(22, 5, 30)
    group3 = np.random.normal(19, 5, 30)

    assumptions = check_anova_assumptions(group1, group2, group3)

    for key, value in assumptions.items():
        print(f"\n{key}:")
        if isinstance(value, dict):
            for k, v in value.items():
                print(f"  {k}: {v}")
        else:
            print(f"  {value}")
```

```python
#24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
from scipy.stats import chi2

def chi_square_variance_test(sample, hypothesized_variance, alpha=0.05, alternative='two-sided'):
    """
    Perform Chi-square test for population variance.

    Parameters:
    - sample: array-like sample data
    - hypothesized_variance: hypothesized population variance (sigma0^2)
    - alpha: significance level (default 0.05)
    - alternative: 'two-sided', 'less', or 'greater'

    Returns:
    - test statistic
    - critical value(s)
    - p-value
    - interpretation string
    """
    n = len(sample)
    sample_var = np.var(sample, ddof=1)  # sample variance with Bessel's correction

    # Test statistic
    chi_stat = (n - 1) * sample_var / hypothesized_variance

    if alternative == 'two-sided':
        # Two-tailed critical values
        cv_lower = chi2.ppf(alpha / 2, df=n - 1)
        cv_upper = chi2.ppf(1 - alpha / 2, df=n - 1)
        p_value = 2 * min(chi2.cdf(chi_stat, df=n - 1), 1 - chi2.cdf(chi_stat, df=n - 1))

        reject = chi_stat < cv_lower or chi_stat > cv_upper
        critical_values = (cv_lower, cv_upper)
        interpretation = ("Reject null hypothesis: Population variance differs from hypothesized variance."
                          if reject else "Fail to reject null hypothesis: No evidence variance differs.")

    elif alternative == 'less':
        # Left-tailed test (variance < hypothesized)
        cv = chi2.ppf(alpha, df=n - 1)
        p_value = chi2.cdf(chi_stat, df=n - 1)
        reject = chi_stat < cv
        critical_values = cv
        interpretation = ("Reject null hypothesis: Population variance is less than hypothesized."
                          if reject else "Fail to reject null hypothesis.")

    else:  # alternative == 'greater'
        # Right-tailed test (variance > hypothesized)
        cv = chi2.ppf(1 - alpha, df=n - 1)
        p_value = 1 - chi2.cdf(chi_stat, df=n - 1)
        reject = chi_stat > cv
        critical_values = cv
        interpretation = ("Reject null hypothesis: Population variance is greater than hypothesized."
                          if reject else "Fail to reject null hypothesis.")

    return chi_stat, critical_values, p_value, interpretation

# Example usage
if __name__ == "__main__":
    np.random.seed(0)
    sample_data = np.random.normal(loc=100, scale=15, size=25)
    hypothesized_var = 225  # 15^2

    chi_stat, crit_vals, p_val, result = chi_square_variance_test(
        sample=sample_data,
        hypothesized_variance=hypothesized_var,
        alpha=0.05,
        alternative='two-sided'
    )

    print(f"Chi-square test statistic: {chi_stat:.4f}")
    print(f"Critical value(s): {crit_vals}")
    print(f"P-value: {p_val:.4f}")
    print(result)

```

```python
#25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

from math import sqrt
from statsmodels.stats.proportion import proportions_ztest

def two_proportion_z_test(count1, nobs1, count2, nobs2, alternative='two-sided'):
    """
    Perform a two-proportion Z-test.

    Parameters:
    - count1: number of successes in group 1
    - nobs1: sample size of group 1
    - count2: number of successes in group 2
    - nobs2: sample size of group 2
    - alternative: 'two-sided', 'smaller', or 'larger'

    Returns:
    - z_stat: Z statistic
    - p_value: p-value for the test
    """
    counts = [count1, count2]
    nobs = [nobs1, nobs2]

    z_stat, p_value = proportions_ztest(count=counts, nobs=nobs, alternative=alternative)

    return z_stat, p_value

if __name__ == "__main__":
    # Example data: successes and sample sizes
    successes_group1 = 45
    size_group1 = 100

    successes_group2 = 30
    size_group2 = 90

    z, p = two_proportion_z_test(successes_group1, size_group1, successes_group2, size_group2, alternative='two-sided')

    print(f"Z-statistic: {z:.4f}")
    print(f"P-value: {p:.4f}")

    alpha = 0.05
    if p < alpha:
        print("Reject the null hypothesis: The proportions are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in proportions.")
```

```python
#26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def f_test_variances(data1, data2, alpha=0.05):
    """
    Perform an F-test to compare variances of two independent samples.

    Parameters:
    - data1, data2: arrays of sample data
    - alpha: significance level (default 0.05)

    Returns:
    - F_stat: calculated F statistic
    - p_value: p-value of the test
    - critical_value_low, critical_value_high: rejection region boundaries
    - interpretation: conclusion string
    """

    n1 = len(data1)
    n2 = len(data2)
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)

    # F-statistic: ratio of variances (larger variance / smaller variance)
    if var1 >= var2:
        F_stat = var1 / var2
        dfn, dfd = n1 - 1, n2 - 1  # degrees of freedom numerator and denominator
    else:
        F_stat = var2 / var1
        dfn, dfd = n2 - 1, n1 - 1

    # Calculate critical values for two-tailed test
    critical_value_low = f.ppf(alpha / 2, dfn, dfd)
    critical_value_high = f.ppf(1 - alpha / 2, dfn, dfd)

    # Calculate p-value for two-tailed test
    if F_stat > 1:
        p_value = 2 * (1 - f.cdf(F_stat, dfn, dfd))
    else:
        p_value = 2 * f.cdf(F_stat, dfn, dfd)

    # Ensure p_value is not greater than 1
    p_value = min(p_value, 1.0)

    # Interpretation
    if p_value < alpha:
        interpretation = "Reject null hypothesis: Variances are significantly different."
    else:
        interpretation = "Fail to reject null hypothesis: No significant difference in variances."

    return F_stat, p_value, critical_value_low, critical_value_high, interpretation, dfn, dfd

def plot_f_distribution(F_stat, critical_low, critical_high, dfn, dfd, alpha=0.05):
    """
    Plot the F-distribution with critical regions and the test statistic.
    """
    x = np.linspace(0, f.ppf(0.999, dfn, dfd), 1000)
    y = f.pdf(x, dfn, dfd)

    plt.figure(figsize=(10,6))
    plt.plot(x, y, label=f'F-distribution (dfn={dfn}, dfd={dfd})')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= critical_low), color='red', alpha=0.3, label='Rejection region (low)')
    plt.fill_between(x, 0, y, where=(x >= critical_high), color='red', alpha=0.3, label='Rejection region (high)')

    # Mark test statistic
    plt.axvline(F_stat, color='blue', linestyle='--', label=f'F statistic = {F_stat:.3f}')

    plt.title('F-Distribution and Test Statistic')
    plt.xlabel('F value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
if __name__ == "__main__":
    np.random.seed(42)
    # Generate two random samples with different variances
    sample1 = np.random.normal(loc=50, scale=5, size=30)   # variance ~ 25
    sample2 = np.random.normal(loc=50, scale=8, size=35)   # variance ~ 64

    F_stat, p_value, crit_low, crit_high, interpretation, dfn, dfd = f_test_variances(sample1, sample2)

    print(f"F-statistic: {F_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical values: low={crit_low:.4f}, high={crit_high:.4f}")
    print(interpretation)

    plot_f_distribution(F_stat, crit_low, crit_high, dfn, dfd)
```

```python
#27.Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy.stats import chisquare

# Simulate observed counts for 4 categories (example)
np.random.seed(0)
observed = np.random.multinomial(n=100, pvals=[0.2, 0.3, 0.25, 0.25])

# Define expected probabilities for the categories
expected_probs = np.array([0.25, 0.25, 0.25, 0.25])
expected_counts = expected_probs * observed.sum()

# Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected_counts)

print(f"Observed counts: {observed}")
print(f"Expected counts: {expected_counts.astype(int)}")
print(f"Chi-square statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution differs significantly from expected.")
else:
    print("Fail to reject the null hypothesis: No significant difference from expected distribution.")
```
